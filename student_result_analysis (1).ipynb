```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Rahul","Aman","Priya","Sneha","Rohit","Anjali","Karan"],
    "Math": [78,65,90,55,72,85,40],
    "Physics": [82,70,88,60,68,80,45],
    "Chemistry": [75,60,92,58,70,84,42]
}

df = pd.DataFrame(data)

df
```

```python
df["Average"] = df[["Math","Physics","Chemistry"]].mean(axis=1)

df
```

```python
top_student = df.loc[df["Average"].idxmax()]

top_student
```

```python
df["Result"] = np.where(df["Average"] >= 50, "Pass", "Fail")

df
```

```python
plt.bar(df["Name"], df["Average"])

plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")

plt.show()
```

```python
subject_avg = df[["Math","Physics","Chemistry"]].mean()

subject_avg
```

```python
subject_avg.plot(kind="bar")

plt.title("Subject Wise Average Marks")
plt.show()
```

```python
print("Top Student:", top_student["Name"])
print("Subject averages:")
print(subject_avg)
```
